### Pregunta 2.1

Un desarrollador del equipo propone: "Para el modelo de radiografías no tiene sentido usar Transfer Learning desde ImageNet; ese dataset tiene fotos de perros, gatos y autos. Las radiografías son imágenes en escala de grises totalmente distintas. Mejor entrenamos desde cero para no contaminar el modelo.”


Con esto responda en su reporte:

**1. ¿Está de acuerdo con el desarrollador? Argumente su posición con base en lo que las capas
convolucionales tempranas de una CNN realmente aprenden y por qué ese conocimiento sí es
transferible, incluso entre dominios visualmente distintos.**

No.
Las primeras capas de una CNN no aprenden “perros o autos”, sino cosas básicas como bordes y formas.
Eso también existe en radiografías, así que sí sirve usar Transfer Learning.

**2. Más allá del argumento técnico, ¿cuál es el costo real de entrenar desde cero para una startup como MediScan Guatemala? Considere al menos dos dimensiones distintas a la puramente
computacional (p. ej., tiempo al mercado, riesgo, talento humano).**

- Más tiempo para tener un buen modelo
- Más riesgo de que falle (porque hay pocos datos)
- Más dificultad técnica (necesita más experiencia)

**3. ¿Existe algún escenario legítimo en el que usted sí consideraría entrenar desde cero en lugar de hacer Transfer Learning? Descríbalo y justifíquelo.**

- Si tienes muchísimos datos
- Si el problema es muy diferente
- Si el modelo preentrenado no ayuda
